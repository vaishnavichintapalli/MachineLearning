# Training and Testing data

```python
x=income.iloc[:,[10,11,12]].values
y=income.iloc[:,[14]].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# Random Forest

```python
#build random forest classifier

from sklearn.ensemble import RandomForestClassifier
rmf=RandomForestClassifier(max_depth=3,random_state=0)
rf_classi=RandomForestClassifier()
rf_classi.fit(x_train,y_train)
```

```python
rf_classi_y_test=rf_classi.predict(x_test)

rf_classi_y_test
```

```python
from sklearn.metrics import confusion_matrix
rf_cm_test=confusion_matrix(y_test,rf_classi_y_test)
rf_cm_test
```

```python
from sklearn.metrics import recall_score
recall_score(y_test,rf_classi_y_test)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test,rf_classi_y_test)
```

```python
#Accuracy score on test data

from sklearn.metrics import accuracy_score
rf_accu_test=accuracy_score(y_test,rf_classi_y_test)
rf_accu_test
```
